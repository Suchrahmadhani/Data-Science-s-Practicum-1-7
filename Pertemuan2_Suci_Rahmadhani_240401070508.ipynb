# Eksplorasi Dataset Titanic

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.info()

df.describe()
```

# Operasi Dasar Pandas

```python
# ==========================================
# Boolean Indexing
# ==========================================

penumpang_selamat = df[df['Survived'] == 1]

penumpang_wanita = df[df['Sex'] == 'female']

wanita_kelas1 = df[(df['Sex']=='female') & (df['Pclass']==1)]

df[['Name','Age','Sex','Survived']]
```

```python
# ==========================================
# GroupBy & Agregasi
# ==========================================

df.groupby('Pclass')['Survived'].mean()

hasil = df.groupby('Pclass').agg(
    total=('PassengerId','count'),
    selamat=('Survived','sum'),
    pct_selamat=('Survived','mean'),
    rata_usia=('Age','mean'),
    max_fare=('Fare','max')
)

hasil = hasil.round(3)

print(hasil)

df.groupby(['Pclass','Sex'])['Survived'] \
    .mean() \
    .unstack() \
    .round(3)
```

# Hands-on Dataset Titanic (Q1–Q6)

# Q1. Struktur Dataset Titanic

```python
# Q1
# Menampilkan jumlah baris, kolom, dan daftar nama kolom

print(f"Jumlah baris dan kolom: {df.shape}")

print("\nDaftar Kolom:")
print(df.columns.tolist())
```

### Interpretasi

Berdasarkan hasil di atas, dataset Titanic terdiri dari **891 baris** dan **12 kolom**. Setiap kolom menyimpan informasi yang berbeda mengenai identitas penumpang, kelas, jenis kelamin, usia, tarif, hingga status keselamatan. Informasi tersebut akan digunakan sebagai dasar untuk analisis pada pertanyaan berikutnya.

# Q2. Missing Values pada Dataset

```python
# Q2
# Menampilkan jumlah dan persentase missing values pada setiap kolom

missing = df.isnull().sum()

persentase = (missing / len(df) * 100).round(2)

missing_df = pd.DataFrame({
    'Jumlah Missing': missing,
    'Persentase (%)': persentase
})

missing_df[missing_df['Jumlah Missing'] > 0]
```

### Interpretasi

Hasil menunjukkan bahwa tidak semua kolom memiliki data yang lengkap. Kolom **Cabin** memiliki jumlah missing value paling banyak sehingga informasi kabin penumpang tidak tersedia untuk sebagian besar data. Selain itu, kolom **Age** juga memiliki cukup banyak data yang kosong, sedangkan kolom **Embarked** hanya memiliki sedikit missing value. Informasi ini penting karena data yang hilang perlu dipertimbangkan sebelum dilakukan analisis lebih lanjut.

# Q3. Persentase Penumpang yang Selamat

```python
# Q3
# Menghitung persentase penumpang yang selamat

persentase_selamat = (df['Survived'].mean() * 100)

print(f"Persentase penumpang yang selamat: {persentase_selamat:.2f}%")
```

### Interpretasi

Berdasarkan hasil perhitungan, sekitar **38%** penumpang berhasil selamat dari kecelakaan Titanic. Artinya, jumlah penumpang yang tidak selamat masih jauh lebih banyak dibandingkan dengan yang selamat. Nilai ini memberikan gambaran awal mengenai kondisi keselamatan penumpang pada saat peristiwa tersebut terjadi.

# Q4. Penumpang Wanita Kelas 1

```python
# Q4
# Menampilkan jumlah dan rata-rata usia penumpang wanita kelas 1

wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

print("Jumlah penumpang wanita kelas 1 :", len(wanita_kelas1))
print("Rata-rata usia :", round(wanita_kelas1['Age'].mean(), 2), "tahun")
```

### Interpretasi

Dari hasil filtering diketahui bahwa terdapat **94** penumpang wanita yang berada di kelas 1. Rata-rata usia mereka sekitar **34,61 tahun**. Hasil ini menunjukkan bagaimana Pandas dapat digunakan untuk memilih data berdasarkan lebih dari satu kondisi sekaligus.

# Q5. Tingkat Keselamatan Berdasarkan Kelas Penumpang

```python
# Q5
# Menghitung tingkat keselamatan berdasarkan kelas penumpang

survival_rate = df.groupby('Pclass')['Survived'].mean()

print(survival_rate)

kelas_tertinggi = survival_rate.idxmax()

print(f"\nKelas dengan tingkat keselamatan tertinggi adalah Kelas {kelas_tertinggi}.")
```

### Interpretasi

Hasil analisis menunjukkan bahwa penumpang **kelas 1** memiliki tingkat keselamatan paling tinggi dibandingkan kelas lainnya. Sementara itu, tingkat keselamatan penumpang kelas 3 merupakan yang paling rendah. Hal ini menunjukkan bahwa kelas penumpang kemungkinan memiliki hubungan dengan peluang untuk selamat.

# Q6. Analisis Tambahan

```python
# Q6
# Analisis tingkat keselamatan berdasarkan jenis kelamin

survival_gender = df.groupby('Sex')['Survived'].mean().round(3)

print(survival_gender)
```

### Interpretasi

Berdasarkan hasil analisis, tingkat keselamatan penumpang wanita jauh lebih tinggi dibandingkan penumpang pria. Hal ini menunjukkan bahwa jenis kelamin menjadi salah satu faktor yang berpengaruh terhadap peluang keselamatan penumpang Titanic. Dari hasil ini saya dapat melihat bagaimana fungsi **groupby()** dapat digunakan untuk membandingkan suatu variabel dengan lebih mudah.
